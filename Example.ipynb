```python
import IASIpy.IASIpy as iasipy
import xarray as xr
import matplotlib.pyplot as plt
```

# read and visualise

iasipy.iasi_gridding(xarray_file, var, var_dim, interval_HMMSS, n_levs_start, n_levs_end)

1. xarray file: input data should be opened through xarray


2. var: variable to be extracted from the dataset


3. var_dim : there are 2 options: 1. profile and 2. column. Eg: if your var is ozone_total_column should be "column", whereas if var is ozone_partial_column_profile, then var_dim should be "profile".


4. interval_HHMMSS: The time interval required to average the swath data.


5. n_levs_start : lowermost vertical layer required


6. n_levs_end : uppermost vertical layer required


```python
#PROFILE

xarray_file = xr.open_dataset('/home/user/Desktop/TROP_OZONE/IASI/IASI_FORLI_O3_metopa_20180103_v20151001.nc')

var = 'ozone_partial_column_profile'

var_dim = 'profile'

interval_HMMSS = 120000 # 2 time axis for a day,(60000 gives 4 time axis a day)

n_levs_start = 0

n_levs_end = 2

data = iasipy.iasi_gridding(xarray_file, var, var_dim, interval_HMMSS, n_levs_start, n_levs_end)
```

```python
data[0,0].plot(cmap='jet', robust=True)
plt.show()

data[0,1].plot(cmap='jet', robust=True)
plt.show()
```

```python
#COLUMN

xarray_file = xr.open_dataset('/home/user/Desktop/TROP_OZONE/IASI/IASI_FORLI_O3_metopa_20180103_v20151001.nc')

var = 'ozone_total_column'

var_dim = 'column'

interval_HMMSS = 120000 # 2 time axis for a day,(60000 gives 4 time axis a day)

n_levs_start = 0  #any value

n_levs_end = 2  ####any value

data = iasipy.iasi_gridding(xarray_file, var, var_dim, interval_HMMSS, n_levs_start, n_levs_end)
```

2. path: the path to iasi files eg. /home/user/Desktop/TROP_OZONE/IASI/
